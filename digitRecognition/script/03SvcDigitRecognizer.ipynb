# Prediction of Digits Using SVC from Scikit-Learn.

## Imports:

```python
import numpy as np 
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score 
import sys
```

## Load Data

```python
data = np.loadtxt(r"D:\Professional\University\DataScience\code\Kaggle\01DigitRecognizer\data\train.csv",skiprows=1, delimiter=',')
```

### Seperate data into X and Y

```python
yTrain = data[:,0]
xTrain = data[:,1:]
```

```python
# Verifying Dataset
data.shape ,xTrain.shape, yTrain.shape 
```

## Modeling

### Load Support Vector Classification Model

```python
model = SVC()
```

### Cross Validate the model with features and output

```python
# cross_val_score(model,xTrain,yTrain)
```

### Fit the model

```python
model.fit(xTrain,yTrain)
```

## Predict.

### Predict with trainning data set.

```python
predictXTrain = model.predict(xTrain)
```

```python
# # Accuracy Percentile
TrainAccuracyPercent = 100*np.sum(predictXTrain==yTrain)/len(yTrain)
TrainAccuracyPercent
```

### Predict test.csv

```python
# Read csv
testData = np.loadtxt(r"D:\Professional\University\DataScience\code\Kaggle\01DigitRecognizer\data\test.csv",skiprows=1, delimiter=',')
testData.shape
```

```python
predictXTest = model.predict(testData)
```
